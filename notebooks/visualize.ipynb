```python
from ultralytics.utils.torch_utils import init_seeds

# To ensure reproducibility
init_seeds(seed=729, deterministic=True)

model_path = '../models/backbone-freeze/best.pt'

# Load model
model = YOLO(model_path)

# Change this to desired save path
# Or leave it as it is
project_name = "Recycle-Trash-3"
val_save_dir = f'../val/{project_name}/runs'

run_name = "test-backbone-freeze" # This will be the child folder of val_save_dir

# Run testing
metrics = model.val(
    data='../config/data.yaml',   # path to data.yaml
    plots=True,
    split='test',
    project=val_save_dir,
    name=run_name)
```

### Confusion Matrix

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# path to validation run directory (change this accordingly)
test_run_dir = '../val/Recycle-Trash-3/runs'

# Get the confusion matrix
confusion_matrix_path = f'{test_run_dir}/confusion_matrix.png'

# Display
img = mpimg.imread(confusion_matrix_path)
plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis('off')
plt.title('Confusion Matrix - Backbone Freeze (first 10 layer)')
plt.savefig('backbone-freeze-confusion-matrix', dpi=300)
plt.show()
```
